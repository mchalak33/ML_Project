# Different Types of Plots in Matplotlib

# Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points = np.array([4,7,2,10,13])
```

```python
plt.plot(points)
plt.show()
```

```python
a2 = np.array([1,2,3,4,5,6])
plt.plot(a2,linestyle ='dashed')
plt.show
```

# Bar Chart

```python
X = np.array(["X","Y","Z","W"])
Y = np.array([25,30,27,22])
```

```python
plt.bar(X,Y)
plt.show()
```

# Scatter Plot

```python
x1=np.array([20,15,11,7,8,10])
y1=np.array([2,7,15,8,10,12])
```

```python
plt.scatter(x1,y1)
plt.show()
```

# Pie Chart

```python
p = np.array([30,20,25,20])
plt.pie(p)
```

# Histogram

```python
h1=np.random.normal(170,10,250)
```

```python
plt.hist(h1)
```
